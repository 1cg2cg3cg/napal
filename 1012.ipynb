### 댓글 분석(명사 태깅)

```python
cnt = collections.defaultdict(int)

kkma = Kkma()
okt = Okt()
for review in reviews[0] :
    
#     print(review['REVIEW'])
#     pprint(kkma.nouns(review['REVIEW']))
#     pprint(okt.pos(review['REVIEW']))
    temp = collections.Counter(okt.pos(review['REVIEW']))
    for t in temp :
        if t[1] == 'Noun' :
#             print(t)
            cnt[t[0]] += 1

# print(cnt)
# print(sorted(cnt, key = lambda x : cnt[x], reverse=True))

# print(sorted(cnt.items(), key = lambda x : x[1], reverse=True))

cnt_nouns = sorted(cnt.items(), key = lambda x : x[1], reverse=True)
# print(cnt_nouns)

noun_list = []
value_list = []

for key, value in cnt_nouns :
    noun_list.append(key)
    value_list.append(value)
# for i in sorted(cnt, key = lambda x : cnt[x], reverse=True) :
#     print(i)

print(noun_list)
print(value_list)
```
